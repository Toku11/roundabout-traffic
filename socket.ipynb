```python
import socket
import time

HOST = '127.0.0.1'  # Standard loopback interface address (localhost)
PORT = 65432        # Port to listen on (non-privileged ports are > 1023)
import numpy as np 
with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    s.bind((HOST, PORT))
    s.listen()
    conn, addr = s.accept()
    val = np.array([1,1,1,1])
    with conn:
        print('Connected by', addr)
        startTime = time.time()
        while True:
            startTime = time.time()
            val +=1
            conn.sendall(f'{val[0]} {val[1]} {val[2]} {val[3]}\n'.encode())
            print(val)
            data = conn.recv(1024)
            if not data:
                break
            print(data, "\n")
```
